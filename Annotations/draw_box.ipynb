```python
import cv2
import pandas as pd
from PIL import Image
```

```python
full_labels = pd.read_csv(r'C:\Users\aurel\Desktop\video_oiseaux_cam2\Data\bird_labels.csv')
full_labels.head()
```

```python
def draw_boxes(image_name):
    selected_value = full_labels[full_labels.filename == image_name]
    img = cv2.imread('D:\\Aurelie\\cam2\\TLC00020\\{}'.format(image_name))
    for index, row in selected_value.iterrows():
        img = cv2.rectangle(img, (row['xmin'], row['ymin']), (row['xmax'], row['ymax']), (0, 255, 0), 3)
    return img
```

```python
Image.fromarray(draw_boxes('C2_TLC00020_21-05-12_18h41.png'))
```
